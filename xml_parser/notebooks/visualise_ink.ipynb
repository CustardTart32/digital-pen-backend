```python
import json 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
```

```python
with open("../data/uploaded_ink/ink.json") as f:
    ink = json.load(f)
```

```python
plt.plot(ink["x"], ink["y"])
plt.gca().invert_yaxis()
plt.show()
```

```python
X = np.swapaxes(np.array([ink["x"], ink["y"], ink["t"]]), 0 , 1) 
```

```python
X_norm = normalize(X, axis=0)
```

```python
kmeans = KMeans(n_clusters=3).fit(X_norm)
labels = kmeans.labels_
```

```python
u_labels = np.unique(kmeans.labels_)

for i in u_labels:
    plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i, s = 1)
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```

```python
X = np.swapaxes(np.array([ink["x"], ink["y"], ink["t"]]), 0 , 1) 
X_norm = normalize(X, axis=0)
```

```python
from sklearn.cluster import DBSCAN

clustering = DBSCAN(eps=0.007, min_samples=5).fit(X_norm)
labels = clustering.labels_
```

```python
u_labels = np.unique(clustering.labels_)

for i in u_labels:
    plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i, s = 1)
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```
